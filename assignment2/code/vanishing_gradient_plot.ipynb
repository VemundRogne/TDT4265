```python
import numpy as np
```

```python
def sech(x):
    return 1 / np.cosh(x)

def lecun_sigmoid_derivative(x):
    return 1.14393 * sech((2 * x)/3) ** 2
```

```python
# Plot the derivative of the sigmoid
import matplotlib.pyplot as plt

x = np.linspace(-3, 3, 100)
y = lecun_sigmoid_derivative(x)

plt.figure()
plt.plot(x, y)
plt.grid()
plt.title("Derivative of improved sigmoid")
plt.show()
```
